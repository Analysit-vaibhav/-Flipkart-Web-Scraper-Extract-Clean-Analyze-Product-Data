```python
from bs4 import BeautifulSoup #importing BeautifulSoup
import requests               # importing requests 
import pandas as pd
import numpy as np
```

```python
# Function for Title
def get_Product(soup):
    try:
        Product = soup.find("span",attrs={"class":"VU-ZEz"}).text.strip()
        return Product
    except AttributeError:
        Product = ""
        return Product
        
# Function for MRP
def get_MRP(soup):
    try:
        MRP = soup.find("div",attrs={"class":"yRaY8j A6+E6v"}).text.strip()
        return MRP
    except AttributeError:
        MRP = ""
        return MRP
        
# Function for Sales Price        
def get_Sales_Price(soup):
    try:
        Sales_Price = soup.find("div",attrs={"class":"Nx9bqj CxhGGd"}).text.strip()
        return Sales_Price
    except AttributeError:
        Sales_Price = ""
        return Sales_Price
        
# Function for Rating        
def get_Rating(soup):
    try:
        Rating = soup.find("div",attrs = {"class":"XQDdHH"}).text.strip()
        return Rating
    except AttributeError:
        Rating = ""
        return Rating

# Function for Product Features
def get_Feature(soup):
    try:
        Features = soup.find("li",attrs = {"class":"_7eSDEz"}).text.strip()
        return Features
    except AttributeError:
        Features = ""
        return Features
```

```python
if __name__ == "__main__":
    # adding user agent
    headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/132.0.0.0 Safari/537.36 Edg/132.0.0.0",
    "X-Amzn-Trace-Id": "Root=1-67a30b22-215f0f4c578a51b94eddcc81"
}


    # webpage url
    url="https://www.flipkart.com/search?q=phone+under+20000&as=on&as-show=on&otracker=AS_Query_OrganicAutoSuggest_5_10_sc_na_na&otracker1=AS_Query_OrganicAutoSuggest_5_10_sc_na_na&as-pos=5&as-type=RECENT&suggestionId=phone+under+20000&requestId=91fd4033-15db-4f68-87f7-15b29dcea5dd&as-searchtext=phone+under+20000&p%5B%5D=facets.brand%255B%255D%3DApple&p%5B%5D=facets.brand%255B%255D%3DSAMSUNG&p%5B%5D=facets.brand%255B%255D%3DGoogle&p%5B%5D=facets.brand%255B%255D%3DMOTOROLA&p%5B%5D=facets.brand%255B%255D%3Dvivo&p%5B%5D=facets.brand%255B%255D%3DOPPO&p%5B%5D=facets.brand%255B%255D%3DREDMI&p%5B%5D=facets.brand%255B%255D%3Drealme&p%5B%5D=facets.brand%255B%255D%3DPOCO&p%5B%5D=facets.brand%255B%255D%3DNothing&p%5B%5D=facets.brand%255B%255D%3DOnePlus&p%5B%5D=facets.brand%255B%255D%3DIQOO&p%5B%5D=facets.rating%255B%255D%3D4%25E2%2598%2585%2B%2526%2Babove"

    # HTTP request
    mainpage = requests.get(url,headers=headers)
    
    # Soup Objectt
    soup = BeautifulSoup(mainpage.content,"html.parser")

    # Fetching all products links
    # links = soup.find_all("a", attrs = {"class":"DByuf4"})
    links = soup.find_all("a", class_="CGtC98")
    
    link_list = []

    for link in links:
        link_list.append(link.get("href"))

    d = {"Product":[],
         "MRP":[],
         "Sales_Price":[],
         "Rating":[],
         "Features":[] 
        }

    # Loop for extracting information for the products

    for link in link_list:
        webpage = requests.get("https://www.flipkart.com/" + link)

        new_soup = BeautifulSoup(webpage.content,"html.parser")


     # calling function to extact all the nessesery info of the product

        d["Product"].append(get_Product(new_soup))
        d["MRP"].append(get_MRP(new_soup))
        d["Sales_Price"].append(get_Sales_Price(new_soup))
        d["Rating"].append(get_Rating(new_soup))
        d["Features"].append(get_Feature(new_soup))


    data = pd.DataFrame.from_dict(d)
    data = data.dropna(subset=["Product"])
    data.to_csv("scraped_data.csv",header=True,index=False)

    
```
